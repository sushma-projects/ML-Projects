```python
def cleanhtml(sentence):
    cleanr = stdre.compile('[<|.|*|?|>]')
    cleantext = stdre.sub(cleanr,' ',sentence)
    return cleantext
```

```python
cleanhtml(data.Massage[690])
```

```python
def cleanpunc(sentence):
    cleaned = stdre.sub('[?|!|\'|"|#]',r'',sentence)
    cleaned = stdre.sub('[.|,|)|(|\|/]',r'',cleaned)
    cleaned = stdre.sub('[<|.|*|?|>]',r'',sentence)
    return cleaned
```

```python
i = 0
str1 = ' '
final_string = []
for sent in data.Massage.values:
    filtered_sentence = []
    sent = cleanhtml(sent)
    for w in sent.split():
        for cleaned_words in cleanpunc(w).split():
            if ((cleaned_words.isalpha()) & (len(cleaned_words)>2)):
                if (cleaned_words.lower()not in stop):
                    s = (sno.stem(cleaned_words.lower()))
                    print(s)
                    filtered_sentence.append(s)
                else :
                    continue
            else:
                continue
    str1 = " ".join(filtered_sentence)
    final_string.append(str1)
    i += 1

    data.Massage = final_string
```
